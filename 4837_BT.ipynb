## BackTracking

```python


def dfs(n, cnt, s):
    
    global ans
    
    # 3. 가지 치기 Pruning, 가능한 모든 경우 처리 후, 시간 초과를 고려하여 처리
    # 음수 가 없으므로, s > K 
    if s > K or cnt > total:   
        return

    # 1. 종료 조건 n에 관련된 수식 
    if n == N:
        
        # 정답 처리 
        if cnt == total and s == K :
            
            ans += 1
        
        return
    
    
    # 2. 하부 함수 호출
    
    dfs(n+1, cnt+1, s+lst[n])  # 현재 n을 사용하는 경우, 다음 n으로 진행, 개수 ++ , 및 합 계산 
    
    dfs(n+1, cnt, s)           # 현재 n을 사용하지 않는 경우, 다음 n으로 진행
    
    
```

```python
T = int(input())

for t in range(1, T + 1):
    
    total, K = map(int, input().split())
    
    N = 12
    
    lst = list(range(1, N+1))
    
    ans = 0
    
    # 초기 상태 호출
    dfs(0, 0, 0)
    
    print('#{0} {1}'.format(t, ans))
    
```

## Combinations

```python
from itertools import combinations
```

```python
T = int(input())

for t in range(1, T + 1):
    
    total, K = map(int, input().split())
    
    lst = list(range(1, 12+1))
    
    N = 12

    ans = 0
    
    for tup in combinations(lst, total):
        
        if sum(tup)== K:
            ans += 1

    
    print('#{0} {1}'.format(t, ans))
```
