## Memo

- 열 트리, 이때 열은 연산자, 연산 가능 횟수

- 연산자 우선순위 x, 앞에서 부터 계산

- n은 연산 대상 idx, n == N, return 종료

- 열 트리로, 모든 연산 수행하도록, 

- 조건 처리를 dfs 함수 parameter에 연산 가능 횟수를 계속 넘겨줘서, 사용 가능 횟수 처리

---

- 연산 중의 값은 -100,000,000 이상 100,000,000 이하
    - `ans_min = int(1e8)`
    - `ans_max = int(-1e8)`

---

```python
# parameter에 연산 가능 횟수를 넘겨주면서 조건 처리 

def dfs(n, num, add, sub, mul, div):
    
    global ans_min
    global ans_max
    
    # 1. 종료 조건 n은 연산대상 
    if n == N:
        
        ans_min = min(ans_min, num)
        
        ans_max = max(ans_max, num)
        
        return
    
    # 2. 하부 함수 처리, 횟수가 0이 아니라면 모두 처리
    if add: 
        dfs(n+1, num + lst[n], add-1, sub, mul, div)
    
    if sub: 
        dfs(n+1, num - lst[n], add, sub-1, mul, div)
    
    if mul: 
        dfs(n+1, num * lst[n], add, sub, mul-1, div)
    
    if div: 
        dfs(n+1, int(num / lst[n]), add, sub, mul, div-1)
 
```

---

```python
T = int(input())

for t in range(1, T+1):

    N = int(input())
    
    add, sub, mul, div = map(int, input().split())
    
    lst = list(map(int, input().split()))
    
    # -100,000,000 이상 100,000,000 이하 처리
    
    ans_min = int(1e8)
    ans_max = int(-1e8)
    
    # n은 연산 대상 lst의 idx
    # 시작 idx 포함
    # 입력 받은 연산 횟수를 parameter 로 처리
    
    dfs(1, lst[0], add, sub, mul, div)
    
    print(f'#{t} {ans_max - ans_min}')
```
