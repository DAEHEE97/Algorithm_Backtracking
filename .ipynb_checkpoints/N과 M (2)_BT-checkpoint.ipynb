# N과 M (2)_BT

- 백트래킹

---

- n : 선택 숫자 개수

- 0 부터 시작해서, M개 고르면 종료


```python
def dfs(n, start, lst):         # 멀티트리
    
    if n == M:                  # M개 선택 완료
        
        ans.append(lst)
        
        return

    for s in range(start, N+1):
        
        dfs(n+1, s+1, lst+[s])
        
```

```python
N, M = map(int, input().split())

ans = []

dfs(0, 1, [])   # 멀티트리

for lst in ans:
    print(*lst)
```

---

---

- 이진 트리 선택 o 선택 x
- n 은 선택할 숫자 (1 ~ N 포함) 

```python
def dfs(n, lst):        # 이진트리
    
    
    # 1. 종료조건, 정답 처리, n은 1~N 포함 선택 O X
    # 정답처리
    if n > N:             
        
        if len(lst) == M: 
            ans.append(lst)
        
        return

    # 2. 하부 함수
    dfs(n+1, lst+[n])   # 선택하는 경우
    
    dfs(n+1, lst)       # 선택하지 않는 경우
```

```python
N, M = map(int, input().split())

ans = []

dfs(1, [])    # 이진트리

for lst in ans:
    print(*lst)
```

---

```python
from itertools import permutations
```

```python
for tup in permutations(lst, M):
    
    print(*tup)
```

---

```python
from itertools import combinations
```

```python
for tup in combinations(lst, M):
    
    print(*tup)
    
```
