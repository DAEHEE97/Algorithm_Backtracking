## Memo

- n은 사용된 숫자의 개수,  7자리 숫자 가 되면 종료

- 1 si,sj n = 1부터 시작

- 4×4 , 여섯 번 이동

- 만들 수 있는 서로 다른 일곱 자리 수들의 개수를 구하는 프로그램, cnt

- 열 트리, 이때 열은 방향으로 설정, 이진트리 x

- 종료조건, 1회시 마다 cnt, cnt n = cnt

- 범위내 이동 

---

```python
# dfs로는 가능한 모든 경우 처리 후, 종료 결과를, 전역변수 sset에 저장
# ans = len(sset), 중복하지 않는 갯수 

def dfs(n, string, ci, cj):
    
    # 1. 종료조건, 정답처리, set에 string을 추가후, main에서 len 처리 
    if n == 7:
        sset.add(string)
        return 
    
    
    # 2. 하부 처리, 4방향, 범위내 가능한 모든 경우 처리
    for di,dj in ((1,0),(-1,0),(0,1),(0,-1)):
        
        ni,nj = ci+di, cj+dj
            
        if 0<=ni<N and 0<=nj<N:
            
            dfs(n+1, string+arr[ni][nj], ni, nj)
    
    
    
```

```python
T = int(input())


for t in range(1, T+1):
    
    N = 4
    arr = [ list(input().split()) for _ in range(N) ]    
    
    sset = set()
    

    # 가능한 모든 위치에서 시작
    
    for si in range(0, N):
        
        for sj in range(0, N):
            
            dfs(1, arr[si][sj], si, sj)
    
    print(f'#{t} {len(sset)}')
    
```
