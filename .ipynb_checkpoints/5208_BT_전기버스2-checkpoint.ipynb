## Memo

- 교체 vs 충전

- 최소한의 교체 횟수, (현재량을 모두 사용해야함, 남은량이 있다면 반드시 사용해야함.)

- 1. 교체 o, 2. 교체 x, 
    - 교체 o, 만약 교체 했는데, 도착 하면 끝
    - 교체 x. 조건 이 있음, 남은량이 있어야, 교체 하지 말고 이동이 가능함.

- 단, 출발지에서의 배터리 장착은 교환횟수에서 제외

- 현재 정류장에서 충전지를 확인했을떄, 종료지점 까지 갈수 있다면, 충전 하고 거기서 종료

---

```python
def dfs(n, cnt, C):
    
    global ans
    
    # 3. 가지치기
    if cnt >= ans:
        return
    
    # 1. 종료조건
    if n == N:
        ans = min(ans, cnt)
        return

    
    # 2. 하부 함수, 호출 가지치기(최소 cnt 유망한 답이 먼저 나오는 방향으로 호출 )
    # 다음 정류장으로 이동 

    if C > 0:                # 교체하지 않는 경우, 한칸 이동 남아있는 량 - 1, 
        dfs(n+1, cnt, C-1)
    
    
    dfs(n+1, cnt+1, lst[n]-1)    # 교체하는 경우, 한칸 이동, 현재 정류장에서 -1, 
```

```python
T = int(input())

for t in range(1, T+1):
    
    lst = list(map(int, input().split()))
    
    # 정류장의 개수 
    N = lst[0]
    
    # 모든 정류장에서 교체할경우 N-1 정류장에서 모두 충전후, N번째 정류장 도착 
    ans = N - 1     
    
    
    # 1번째 정류장(idx 1) 에서의 배터리 장착은 교환횟수에서 제외
    # 2번째 정류장 (idx 2)부터 확인, 
    
    # 1번 정류장에서 충전 후 출발, 교체 X, 다음 n 2번부터 진행
    dfs(2, 0, lst[1]-1) 
    
    print(f'#{t} {ans}')
```
