# [14899_BT_스타트와 링크](https://www.acmicpc.net/problem/14889)



- 팀을 구성하는, 가능한 모든 경우 처리시, 시간 내 처리 가능, N  - 4 ~ 20 이므로
- 이진트리 구조의 백트래킹


---

- 가능한 모든 경우 처리, 종료조건 사람수 를 모두 이동 처리시, 0번 부터 idx, 
- 정답처리, 정확히 반으로 나뉘어져있어야 종료
- 가지치기 호출을 줄이지만, 테스트케이스나 조건 검사에 의해 시간 감소 효과 없을수 있음.

---

```python
def cal(alst, blst):
    
    asum = 0
    bsum = 0
    
    # 1,1, 1,2, 2,1, 2,2 
    for i in range(0, M):
        
        for j in range(0, M):
            
            asum += arr[alst[i]][alst[j]]
            bsum += arr[blst[i]][blst[j]]
            
    return abs(asum-bsum)
```

```python
def dfs(n, alst, blst):
    
    global ans
    
    
    # 3-1) 가지치기 1
    if len(alst) > M or len(blst) > M:
        
        return
    
    # 3-2) 가지치기 2
    if ans == 0:
        return
    
    
    # 1. 종료조건 n == idx 4, 정답 처리 팀 카운트 or ,    
    if n == N:
        
        if len(alst) == len(blst):
            
            ans = min(ans, cal(alst,blst))
        
        return
    
    # 2. 하부 함수 호출
    
    # 이진트리
    dfs(n+1, alst+[n], blst)
    
    dfs(n+1, alst, blst+[n])
    
```

---

```python
N = int(input())

arr = [ list(map(int, input().split())) for _ in range(N) ]

M = N // 2

ans = 100000
dfs(0, [], [])
print(ans)
```

```python
def dfs(n, alst, blst):
    
    global ans
    
    
    # 3-1) 가지치기 1
    if len(alst) > M or len(blst) > M:
        
        return
    
    # 3-2) 가지치기 2
    if ans == 0:
        return
    
    
    # 1-1) 종료조건 n == idx 4 
    if n == N:
        
        # 1-2) 정답 처리 조건
        if len(alst) == len(blst):
            
                asum = 0
                bsum = 0
                
                for i in range(0, M):
                    for j in range(0, M):

                        asum += arr[alst[i]][alst[j]]
                        bsum += arr[blst[i]][blst[j]]
                        
                ans = min(ans, abs(asum-bsum))
        return
    
    # 2. 하부 함수 호출
    
    # 이진트리
    dfs(n+1, alst+[n], blst)
    
    dfs(n+1, alst, blst+[n])
    
```
