## parameter

- n은 lst사이에 들어갈 idx 이자, 연산 횟수, 

- parameter에 연산 가능 횟수를 넘겨주면서 횟수, 조건 처리 



```python
def dfs(n, result, p, m, mul, div):
    
    global ans_max
    
    global ans_min
    
    
    
    # 0 4, 연산자의 개수, 길이 x
    if n == N - 1:
        
        ans_max = max(ans_max, result)
        ans_min = min(ans_min, result)
        
        return
    
    # dfs 처리 무조건 가능 x
    # dfs 처리 조건, 횟수가 남아 있어야함
        
    if p > 0 :
        dfs( n+1, result+lst[n+1], p-1, m, mul, div)
    
    if m > 0:
        dfs( n+1, result-lst[n+1], p, m-1, mul, div)

    if mul > 0:
        dfs( n+1, result*lst[n+1], p, m, mul-1, div)

    if div > 0:
        dfs( n+1, int(result/lst[n+1]), p, m, mul, div-1)

    
    
```

```python
T = int(input())


for t in range(1, T+1):
    
    N = int(input()) 
    
    # 4개
    cal = list(map(int, input().split()))
    
    # N개 
    lst = list(map(int, input().split()))
    
    
    ans_max = -(100000000) # int(1e8)
    ans_min = 100000000 # int(-1e8)
    
    # n은 0 start 해야 n+1 가능 0
    
    dfs(0, lst[0], cal[0], cal[1], cal[2], cal[3])
    
    ans = ans_max - ans_min
    
    print(f'#{t} {ans}')
    
```
