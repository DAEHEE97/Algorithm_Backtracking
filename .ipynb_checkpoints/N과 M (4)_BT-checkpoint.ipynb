# [N과 M (4)](https://www.acmicpc.net/problem/15652)

---

## 열 트리

- 단 조건, tmp가 1로 시작하면, N 까지, 2로 시작하면, 2부터 N까지
    - 중복처리랑 다름

- start를 변수 처리, n 은, 선택한 숫자 개수


---

```python
def dfs(n, tlst, start):         
    
    # 1. 종료조건, M개 가 되면 종료, 정답처리
    if n == M:                  
        
        ans.append(tlst)
        
        return
    

    # 2. 하부 함수 dfs
    # 다음 단계로 넘어갈때, start는 start부터 값 대입
    # idx 반복하면서 실행 
    
    for j in range(start, N):
        
        dfs(n+1, tlst + [lst[j]], j)
            
```

```python
N, M = map(int, input().split())

lst = list(range(1, N+1))

ans = []

dfs(0, [], 0)   


for l in ans:
    print(*l)
```
