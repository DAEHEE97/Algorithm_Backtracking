## Memo

- 3 ≤ N ≤ 10, 가능한 모든 경우 처리 가능, 열 트리

- n은 2차원 배열의 행, n == 마지막 행 까지 가면 종료

- 방문 가능한 j 열 조건,  단, 세로로 같은 줄에서 두 개 이상의 숫자를 고를 수 없다.
     - v 방문 처리 lst 생성, 방문 표시, 

- `v[j] = 0 ` state recovery 현재 경로에서의 상태를 이전 상태로 되돌리는 역할, 상태 복원 처리, 

---

```python
def dfs(n, s):
    
    global ans

    # 3. 가지치기, 합이, 이미 최소 값으로 만들어낸 ans 보다 이미 크다면,  
    if s >= ans :
        return    
    
    # 1. 종료 조건, n 은 2차원 배열의 행, 정답 처리 
    if n == N:
        
        ans = min(ans, s) # 정답 처리
        
        return
     
    
    # 2) j 열 트리, N-1 idx 
    # 2-1) v lst, 중복되는 idx 처리
    
    for j in range(0, N):
        
        if v[j] == 0:     # 방문 하지 않은 열 j
            
            v[j] = 1
            
            dfs( n+1, s + arr[n][j] )
            
            v[j] = 0      # state recovery
    
```

```python
T = int(input())

for t in range(1, T+1):
    
    N = int(input())

    arr = [ list(map(int, input().split())) for _ in range(N) ]
    
    v = [0] * N # 방문 처리 lst, j열 용 미방문시 가능

    
    
    # 최소 값 처리, 10보다 작은 자연수가 주어진다. 3≤N≤10
    # 10 * 10 
    
    ans = 10*10
    
    dfs(0, 0)
    
    print(f'#{t} {ans}')
    
```
