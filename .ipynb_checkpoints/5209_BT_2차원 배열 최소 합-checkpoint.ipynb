## 방문 표시 리스트 v [j] = 1, v[j] = 0

```python
def dfs(n, s):
    
    global ans
    
    if s > ans:
        return
     
    if n == N:
        
        ans = min(ans, s)
        return
    
    # j = 0 ~ N-1 까지 
    for j in range(0, N):
        
        if v[j]==0:
            
            v[j] = 1
            
            dfs(n+1, s+arr[n][j])
            
            v[j] = 0
 
```

```python
T = int(input())

for t in range(1, T+1):
    
    N = int(input())
    
    arr = [ list(map(int, input().split())) for _ in range(N) ]
    
    ans = 100 * N
    
    v = [0] * N
    
    dfs(0, 0)
    
    print(f'#{t} {ans}')
    
    
```
