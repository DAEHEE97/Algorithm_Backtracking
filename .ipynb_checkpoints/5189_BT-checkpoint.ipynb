## Memo



- n 은 방문한 구역 수(cnt), 중복 체크 v

- 1 출발, n 은 1부터 start, n이 3개가 되면, 1로 돌아오는걸 합쳐서 종료

- (1,2,3,1) = arr[1][2] + arr[2][3] + arr[3][1]

- (1,3,2,1) = arr[1][3] + arr[3][2] + arr[2][1]

---

```python
def dfs(n, s, ci):
    
    global ans
    
    if s >= ans:
        return
    
    if n == N:
        # 여태까지의 소모량 + 1번으로 복귀비용
        ans = min(ans, s + arr[ci][1])
        
        return
    
    for j in range(2, N+1):
        
        if v[j] == 0:
            
            v[j] = 1
            
            dfs(n+1, s + arr[ci][j], j)
            
            v[j] = 0
```

```python
T = int(input())

for t in range(1, T + 1):
    
    N = int(input())
    
    arr = [ [0] * (N + 1)] + [ [0] + list(map(int, input().split())) for _ in range(N) ]
     
    ans = 100*N
    
    v = [0]*(N+1)
    
    dfs(1, 0, 1)
    
    print(f'#{t} {ans}')
```
