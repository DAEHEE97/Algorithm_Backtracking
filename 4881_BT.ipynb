## Memo

- 어떤 열(숫자)을 선택하는지에 따라서 가능한 모든 경우를 표현

- 가능한 모든 경우 처리 정답 처리 , 상태 공간 트리로, 모든 경우 표시 -> 일반화

- 무조건 선택 가능 x : 단 조건, 세로로 같은 줄에서 중복 x

```python
def dfs(n, s):
    
    global ans

    
    # 3. 가지치기, 합이, 이미 최소 값으로 만들어낸 ans 보다 이미 크다면,  
    if s >= ans :
        return    
    
    # 1. 종료 조건, 가능한 모든 경우의 수 처리 
    if n == N:
        
        ans = min(ans, s) # 정답 처리
        
        return
     
    
    # 2. 무조건 선택 가능 x, 세로로 같은 줄 x, 이진트리 형태 복원 처리
    # 방문 한적 x, dfs를 통해 방문 처리
    
    for j in range(0, N):
        
        if v[j] == 0:     # 사용하지 않은 숫자
            
            v[j] = 1
            
            dfs( n+1, s + arr[n][j] )
            
            v[j] = 0      # 무효화
    
```

```python
T = int(input())

for t in range(1, T + 1):
    
    N = int(input())
    
    arr = [ list(map(int, input().split())) for _ in range(N) ]
    
    ans = N ** 10
    
    v = [0] * N
    
    dfs(0, 0)
    
    print(f'#{t} {ans}')
    
```
