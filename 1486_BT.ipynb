- 조건 : 키의 합 > 선반의 높이 B, 

- 정답 처리 : 그 중 가장 낮은 키의 합 과의 차이 의 최소 s-B


---

## Memo

- 어떤 열(숫자)을 선택하는지에 따라서 가능한 모든 경우, 20**20 케이스 시간 초과 발생

- 2진트리 형태 로 가능하다면, 2진트리로 구현, 2**20 케이스

- 높이가 B 이상인 탑 중에서, 탑의 높이와 B의 차이가 가장 작은 것


---

---

## BackTracking

```python
def dfs(n, s):
    
    global ans
    
    # 3. 가지 치기
    if s - B >= ans:
        return

    # 1. 종료 조건, 정답 처리 ans : 탑의 높이와 B의 차이가 가장 작은 것
    if n == N:
        
        if s >= B:
            ans = min(ans, s - B)
        
        return
    
    # 2. 이진 트리 하부 함수
    dfs(n+1, s+lst[n]) # 포함 o
    
    dfs(n+1, s)        # 포함 x
```

```python
T = int(input())

for t in range(1, T+1):
    
    N, B = map(int, input().split())
    
    lst = list(map(int, input().split()))
    
    ans = N * 10000

    dfs(0, 0)
    
    print(f'#{t} {ans}')
```

## combinations

```python
from itertools import combinations

T = int(input())

for t in range(1, T+1):
    
    N, B = map(int, input().split())
    
    lst = list(map(int, input().split()))
    
    ans = N * 10000

    for n in range(1, N+1):
        
        for tup in combinations(lst, n):
            
            if sum(tup) >= B:
                ans = min(ans, sum(tup) - B)
                
        
    print(f'#{t} {ans}')
```
