## 열 트리

- v 방문 표시 리스트 idx 로 중복 처리 

- 상태 공간 트리 n = 행(i), 가지는 열 (j)

- 모든 가능한 경우를 트리로 생성 후, dfs 조건은 미방문으로 처리

- n = 15, 15 ** 15 케이스, 시간 초과 > 가지 치기 pruning,
    - s 로 처리하는게 시간 유리

```python
def dfs(n, tlst):
    
    global ans

    if sum(tlst) >= ans:
        return
    
    # 열 트리
    # n은 2차원 arr의 행,
    
    if n == N:
        
        ans = min(ans, sum(tlst))
        
        return
    
    # 2) j 열 트리, N idx 
    # 2-1) v lst, 중복되는 idx 처리
    
    for j in range(0, N):
        
        if v[j] == 0:
            
            v[j] = 1
            
            dfs(n+1, tlst+[arr[n][j]])
            
            v[j] = 0
 
```

```python
T = int(input())

for t in range(1, T+1):
    
    N = int(input())
    
    arr = [ list(map(int, input().split())) for _ in range(N) ]
    
    ans = 100 * N
    
    v = [0] * N
    
    dfs(0, [])
    
    print(f'#{t} {ans}')
    
    
```

## 열 트리
- s
- 방문 표시 리스트 v [j] = 1, v[j] = 0

```python
def dfs(n, s):
    
    global ans
    
    if s > ans:
        return
     
    # 열 트리
    # n은 2차원 arr의 행,
    
    if n == N:
        
        ans = min(ans, s)
        return
    
    # 2) j 열 트리, N-1 idx 
    # 2-1) v lst, 중복되는 idx 처리
    
    for j in range(0, N):
        
        if v[j]==0:
            
            v[j] = 1
            
            dfs(n+1, s+arr[n][j])
            
            v[j] = 0
 
```

```python
T = int(input())

for t in range(1, T+1):
    
    N = int(input())
    
    arr = [ list(map(int, input().split())) for _ in range(N) ]
    
    ans = 100 * N
    
    v = [0] * N
    
    dfs(0, 0)
    
    print(f'#{t} {ans}')
    
    
```
