# [N과 M (8)](https://www.acmicpc.net/problem/15657)

---

- 열 트리

- n : 선택한 숫자 개수


---

```python
def dfs(n, tmp, start):         
    
    if n == M:                  
        
        ans.append(tmp)
        
        return

    # idx start 부터, N-1 까지, 0부터  
    # 반복하면서 실행 
    for j in range(start, N):
        
        dfs(n+1, tmp + [lst[j]], j)
            
```

```python
N, M = map(int, input().split())

lst = list(map(int, input().split()))
lst.sort()

ans = []

dfs(0, [], 0)   


for l in ans:
    print(*l)
```

---

```python
def dfs(n, lst, start ):         # 열 트리
    
    if n == M:                  # M개 선택 완료
        
        ans.append(lst)
        
        return

    # 1 부터 1~N 까지 호출
    # 2, 2~N 까지 호출
    for j in range(start, N+1):
        
        dfs(n+1, lst+[j], j)
        
```

```python
N, M = map(int, input().split())

ans = []

dfs(0, [], 1)   # 멀티트리

for lst in ans:
    print(*lst)
```

- n은 숫자 n을 포함할지 안할지
