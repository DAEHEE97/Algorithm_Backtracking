# [N과 M (8)](https://www.acmicpc.net/problem/15657)

---

- 열 트리

- 가능한 모든 경우의 수 처리, (같은 idx j 를 여러번 골라도 됨) - idx 0, idx 0 가능

- 선택 가능한 j 조건, start 변수를 이용해서 처리 

- (중복되는 수열 가능)


---

```python
def dfs(n, tmp, start):         
    
    if n == M:                  
        
        ans.append(tmp)
        
        return

    # 2-1) idx 중복 처리 가능
    # 2-2) start 처리, 다음 턴 j는 start idx j 부터 처리
    
    for j in range(start, N):
        
        dfs(n+1, tmp + [lst[j]], j)
            
```

```python
N, M = map(int, input().split())

lst = list(map(int, input().split()))
lst.sort()

ans = []

dfs(0, [], 0)   


for l in ans:
    print(*l)
```
