- 열 트리

- 가능한 모든 경우의 수 처리, (같은 idx j 를 여러번 골라도 됨) - idx 0, idx 0 가능 

- 중복 되는 수열 불가 처리 - 1,1 고르고, 다음 또 1,1, 불가
    1. set 처리
    2. prev 변수 처리 

- j = 0 부터 시작해서, idx 0 값 이 들어와 있는 상태에서, 이때 다음 lst에 값이 이전 prev 와 같은 값이라면, 처리 x

---

```python
def dfs(n, tmp):
    
    if n == M:
        
        ans.append(tmp)
        
        return
    
    # 2-1) idx 중복 o
    # 2-2) prev, 중복되는 수열 처리 
    
    prev = 0
    for j in range(0, N):
        
        if prev != lst[j]:
            
            prev = lst[j]
            
            dfs( n+1, tmp+[lst[j]] )
            
```

```python
N, M = map(int, input().split())

lst = list(map(int, input().split()))

lst.sort()

ans = []

dfs(0, [])

for l in ans:
    print(*l)
```
