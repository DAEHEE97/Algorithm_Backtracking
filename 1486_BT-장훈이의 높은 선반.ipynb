## 이진 트리

- 가능한 모든 경우 처리, 서로 다른 조합, 이진 트리
- n은 lst의 idx
- 종료 조건, idx 끝까지 실행
- 정답 처리, 특정 값 이상이면서, 차이가 최소가 되도록,
- tlst 처리

```python
def dfs(n, tlst):
    
    global ans
    # n은 idx
    
    if n == N:
        
        if sum(tlst) >= B:
            ans = min(ans, sum(tlst)-B)
        
        return 
    
    
    dfs(n+1, tlst+[lst[n]])
    
    dfs(n+1, tlst)
```

```python
T = int(input())

N, B = map(int, input().split())

lst = list(map(int, input().split()))

ans = 10000

dfs(0, [])

ans
```

## 이진 트리

- s 

```python
def dfs(n, s):
    
    global ans
    
    # 3. 가지 치기
    if s - B >= ans:
        return

    # 1. 종료 조건, 정답 처리 ans : 탑의 높이와 B의 차이가 가장 작은 것
    if n == N:
        
        if s >= B:
            ans = min(ans, s - B)
        
        return
    
    # 2. 이진 트리 하부 함수
    dfs(n+1, s+lst[n]) # 포함 o
    
    dfs(n+1, s)        # 포함 x
```

```python
T = int(input())

for t in range(1, T+1):
    
    N, B = map(int, input().split())
    
    lst = list(map(int, input().split()))
    
    ans = N * 10000

    dfs(0, 0)
    
    print(f'#{t} {ans}')
```

## combinations

```python
from itertools import combinations

T = int(input())

for t in range(1, T+1):
    
    N, B = map(int, input().split())
    
    lst = list(map(int, input().split()))
    
    ans = N * 10000

    for n in range(1, N+1):
        
        for tup in combinations(lst, n):
            
            if sum(tup) >= B:
                ans = min(ans, sum(tup) - B)
                
        
    print(f'#{t} {ans}')
```
