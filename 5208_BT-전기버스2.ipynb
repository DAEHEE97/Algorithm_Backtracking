교체 != 충전

- 교체 o, 남아 있는 량이 0보다 적으면 무조건 교체해야됨
- 교체 x, 남아 있는 량이 0보다 크면, 교체 할 필요 없음

    - 여기서 문제가 교체를 최소로 하는 문제이므로, 교체를 하지 않는 경우를 최우선시 처리

```python
def dfs(n, cnt, r):
    
    global ans
    
    # 시간 초과, 가지치기
    
    if cnt >= ans:
        return
    

    if n == N:
        
        ans = min(ans, cnt)
        return 
    
    # 하부 함수 호출, 특정 조건 없이 가능한가? x, 남은 량이 있어야 교체 x 할수 있음
    # 그냥 교체는 항상 가능 + 남은량이 없으면 무조건 교체 해야됨
    
    # 교체를 최소로 하는 문제이므로, 교체를 하지 않는 경우를 최우선시
    if r > 0:
        dfs(n+1, cnt, r-1)        # 교체 x

    dfs(n+1, cnt+1, lst[n]-1)     # 교체 O

```

```python
T = int(input())

for t in range(1, T+1):
    
    lst = list(map(int, input().split()))
    
    N = lst[0]
        
    ans = N

    # 1 부터 시작, 
    dfs(1, 0, lst[1])

        
    print(f'#{t} {ans}')
    
```
