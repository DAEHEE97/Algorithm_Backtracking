A1, A2, ... , AN의 N개의 자연수가 주어졌을 때, 최소 1개 이상의 수를 선택하여 그 합이 K가 되는 경우의 수를 구하는 프로그램을 작성하시오.


---

## Memo

- 1개 이상을 선택해서, 합이 K가 되는 모든 경우의 수 

- 가능한 모든 경우 처리

- 이진 트리(lst의 idx) 

- n은 lst idx 0,1,2,3,4 , 합 연산에 포함 할지 안할지, 1
    - 리스트의 값들 중 1 개만 선택해도 가능, 2 개만 선택해도 가능, 모두 선택 가능 끝나도 가능

- 이진 트리 형태 

```python
def dfs(n, s):
    
    global ans
    
    # 3. 가지치기 / 자연수의 합 
    if s > K:
        return 
    
    
    
    # 1. dfs 종료조건 lst idx 종료
    if n == N:
        
        if s == K:
            ans += 1
        
        return 
        
    
    
    dfs(n+1, s+lst[n])
    
    dfs(n+1, s)
    
    
```

```python



T = int(input())


for t in range(1, T+1):
    
    N, K = map(int, input().split())
    
    lst = list(map(int, input().split()))
    
    ans = 0 
    
    dfs(0, 0)
    
    print(f'#{t} {ans}')
    
```

## combinations

```python
from itertools import combinations

T = int(input())

for t in range(1, T+1):
    
    N, K = map(int, input().split())
    
    lst = list(map(int, input().split()))
    
    ans = 0
    
    for i in range(1, N+1):
        
        for tup in combinations(lst, i):
            
            if sum(tup) == K:
                ans += 1
                
    
    print(f'#{t} {ans}')
        
        
```
