## 열 트리

- n 은 선택된 방문한 구역수, 길이 (2차원 arr, 무조건 n은 행 아님),


    - j 열 1부터 시작
    
    - 방문 가능 j 열 처리, 중복 x, v lst 
    
    - 행 정보 ci, 가 있어야하며, 현재 j를 parameter로 처리하여, 다음 행 ci < 현재 j 열 처리
    
    - n start 1로 시작, 는 arr[ci][0] 를 따로 계산 하므로, 
    - 종료 N은 N으로 처리

```python
def dfs(n, s, ci):
    
    
    global ans
    
    # n은 선택된 arr의 길이, 갯수 
    
    if n == N:
        
        print(s+arr[ci][0])
        
        ans = min(ans, s+arr[ci][0])
        
        return
    
    # 2-1) 1열 부터 처리, 
    for j in range(1, N):
        
        if v[j] == 0:
            
            v[j] = 1
            
            dfs(n+1, s+arr[ci][j], j)
            
            v[j] = 0
        
        
```

```python
T = int(input())

for t in range(1, T+1):
    
    N = int(input())
    
    arr = [ list(map(int, input().split())) for _ in range(N) ]
     
    ans = 100*N
    
    v = [0] * N
    
    # last 1 추가 상태, s = 0, ci 행 0 부터 시작
    dfs(1, 0, 0)
    
    print(f'#{t} {ans}')
```
